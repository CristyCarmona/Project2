```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file="food.csv"
data= pd.read_csv(file)
data
```

```python
#removal of columns that we won't use from the first csv
food_analysis_df= data[['Category', 'Description', 'Nutrient Data Bank Number','Data.Carbohydrate', 'Data.Cholesterol', 'Data.Fiber', 'Data.Kilocalories', 'Data.Protein', 'Data.Sugar Total', 'Data.Water', 'Data.Fat.Monosaturated Fat', 'Data.Fat.Polysaturated Fat', 'Data.Fat.Saturated Fat', 'Data.Fat.Total Lipid', 'Data.Major Minerals.Copper', 'Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium', 'Data.Major Minerals.Phosphorus','Data.Major Minerals.Potassium', 'Data.Major Minerals.Sodium', 'Data.Major Minerals.Zinc', 'Data.Vitamins.Vitamin A - IU','Data.Vitamins.Vitamin A - RAE','Data.Vitamins.Vitamin B12', 'Data.Vitamins.Vitamin B6', 'Data.Vitamins.Vitamin C', 'Data.Vitamins.Vitamin E', 'Data.Vitamins.Vitamin K']]
food_analysis_df.head()
```

```python
nutrients=pd.DataFrame(list(food_analysis_df.columns.values))
```

```python
nutrients=nutrients[0].str.split(".", expand=True)
```

## Essential Nutrient Group

```python
#Create essential nutrient group table
essential_nutrient_group= nutrients[[1]].copy()
```

```python
essential_nutrient_group= essential_nutrient_group.dropna().reset_index(drop=True)
```

```python
essential_nutrient_group= essential_nutrient_group.rename(columns={1:"group_name"})
```

```python
essential_nutrient_group['group_name'] = essential_nutrient_group['group_name'].replace(
    {'Carbohydrate': 'Others', 'Cholesterol': 'Others','Fiber': 'Others','Kilocalories': 'Others',
    'Protein': 'Others','Sugar Total': 'Others','Water': 'Others'})
```

```python
essential_nutrient_group= essential_nutrient_group.drop_duplicates().reset_index(drop=True)
```

```python
essential_nutrient_group.insert(0,'id_essengroup',range(1,1+len(essential_nutrient_group)))
essential_nutrient_group
```

## Essential Nutrients

```python
essential_nutrients= nutrients[[1,2]].copy()
```

```python
for index, row_value in enumerate(essential_nutrients[2].isnull()):
    if row_value:
         essential_nutrients.at[index, 2] = essential_nutrients.at[index, 1]
```

```python
essential_nutrients= essential_nutrients.dropna()
essential_nutrients.iloc[0:7,0]="Others"
essential_nutrients= essential_nutrients.rename(columns={1:"group_name",2:"nutrient_name"})
essential_nutrients.insert(0,'id_nutrient',range(1,1+len(essential_nutrients)))
essential_nutrients=essential_nutrient_group.merge(essential_nutrients, on="group_name",how='left')
essential_nutrients= essential_nutrients.dropna()
essential_nutrients=essential_nutrients.drop(['group_name'], axis=1)
essential_nutrients= essential_nutrients.reindex(columns=['id_nutrient','id_essengroup','nutrient_name'])
essential_nutrients
```

# Food Category

```python
#Create food category table
food_category= data[["Category"]].copy()
food_category= food_category.drop_duplicates().reset_index(drop=True)
food_category=food_category.rename(columns={"Category":"desc_category"})
food_category.insert(0,'id_category',range(1,1+len(food_category)))
food_category
```

# Food tbl

```python
food= food_analysis_df[["Category","Nutrient Data Bank Number","Description"]].copy()
column=food_analysis_df["Data.Carbohydrate"].copy()
food["nutrient_amount"]=column
food["nutrient"]= "Data.Carbohydrate"
```

```python
iterate=food_analysis_df.columns[3:]

food_tbl=pd.DataFrame()

for column_name in iterate:
    food= food_analysis_df[["Category","Nutrient Data Bank Number","Description"]].copy()
    column=food_analysis_df[column_name].copy()
    food["nutrient_amount"]=column
    food["nutrient"]= column_name
    food_tbl=food_tbl.append(food, ignore_index=True)
    
food_tbl
```

```python
food_tbl=food_tbl.rename(columns={"Category":"desc_category"})
food_tbl=food_tbl.merge(food_category, on="desc_category",how='left')
food_tbl[['Data','Category','nutrient_name']]=food_tbl['nutrient'].str.split('.',expand=True)
food_tbl
```

```python
for index, row_value in enumerate(food_tbl['nutrient_name'].isnull()):
    if row_value:
         food_tbl.at[index, "nutrient_name"] = food_tbl.at[index, "Category"]
```

```python
food_tbl=food_tbl.merge(essential_nutrients, on="nutrient_name",how='left')
food_tbl=food_tbl.rename(columns={"Nutrient Data Bank Number":"id_food","Description":"food_desc"})
food_tbl = food_tbl.drop(['desc_category','nutrient','Data','Category','nutrient_name','id_essengroup'], axis=1)
food_tbl = food_tbl.reindex(columns=['id_food','food_desc','id_category','id_nutrient','nutrient_amount'])
food_tbl
```

# Create Database Connection

```python
rds_connection_string = "postgres:123456@localhost:5432/food_nutrients"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

# Load dataframes to SQL 

```python
essential_nutrient_group.to_sql(name='essential_nutrient_group', con=engine, if_exists='append', index=False)
```

```python
essential_nutrients.to_sql(name='essential_nutrients', con=engine, if_exists='append', index=False)
```

```python
food_category.to_sql(name='food_category', con=engine, if_exists='append', index=False)
```

```python
food_tbl.to_sql(name='food_tbl', con=engine, if_exists='append', index=False)
```
